```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

Our dataset contains 1000 images generated by each network: `normal`, `thin`, `mnist` and `mostlyconvex`. These images are stored in `shapes.hdf5` file. It can be accessed as shown below:

```python
# Data is grouped by the name of the network:
with h5py.File('shapes.hdf5', 'r') as f:
    for key in f.keys():
        print('%s contains %i images' % (key, len(f[key])))
```

Here we show few images from each group:

```python
def load_images(group):
    data = []
    with h5py.File('shapes.hdf5', 'r') as f:
        for i in range(4):
            data = data + [np.array(f['%s/%i' % (group, i)])]
    return data

def show_images(data):
    fig = plt.figure()
    for (i, img) in enumerate(data):
        ax = fig.add_subplot(2, 2, i+1)
        ax.imshow(img)
```

```python
show_images(load_images('normal'))
```

```python
show_images(load_images('thin'))
```

```python
show_images(load_images('mnist'))
```

```python
show_images(load_images('mostlyconvex'))
```
