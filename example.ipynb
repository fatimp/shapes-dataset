```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

Our dataset contains 1000 images generated by each network: `normal`, `thin`, `mnist` and `mostlyconvex`. These images are stored in `shapes.hdf5` file. It can be accessed as shown below:

```python
# Data is grouped by the name of the network:
with h5py.File('shapes.hdf5', 'r') as f:
    for key in f.keys():
        print('%s contains %i images' % (key, len(f[key])))
```

```python
# Here we show one image from each group:
with h5py.File('shapes.hdf5', 'r') as f:
    normal = np.array(f['normal/0'])
    thin   = np.array(f['thin/0'])
    mnist  = np.array(f['mnist/0'])
    mc     = np.array(f['mostlyconvex/0'])
```

```python
fig = plt.figure()
ax = fig.add_subplot(2, 2, 1)
ax.imshow(normal)
ax = fig.add_subplot(2, 2, 2)
ax.imshow(thin)
ax = fig.add_subplot(2, 2, 3)
ax.imshow(mnist)
ax = fig.add_subplot(2, 2, 4)
ax.imshow(mc)
```
